```python
import pandas as pd
```

```python
data=pd.read_table("/content/sample_data/data.txt", ",")
data.drop_duplicates(subset ="ProductName",keep = False, inplace = True)
data.sort_values(by="ProductPrice",inplace=True)
```

```python
!pip install plotly==4.8.2
import plotly.express as px
```

```python
data.describe()
```

## Pie Chart

```python
visualizedData=px.pie(data, values='ProductPrice', names='ProductBrand',title='Vatan Computer Phones Market Sharing',color_discrete_sequence=px.colors.qualitative.Bold)
visualizedData.show()
```

```python
groupedData=data.groupby('ProductBrand').count().reset_index()
visualizedData=px.pie(groupedData, values='ProductName',labels={"ProductName":"Count"}, names='ProductBrand',title='Vatan Computer Phone Count Sharing',color_discrete_sequence=px.colors.qualitative.Bold)
visualizedData.show()
```

# Histogram

```python
histo=px.histogram(data,x="ProductBrand",y="ProductPrice")
histo.show()
```

# Bubble Charts

```python
newFig=px.strip(data,x="ProductBrand",y="ProductPrice")
newFig.show()
```

```python
visualizedData = px.scatter(data, x="ProductBrand", y="ProductPrice",size="ProductRate")
visualizedData.show()
```

# Sunburst

```python
newFig2=px.sunburst(data,values="ProductPrice",path=["ProductBrand","ProductPrice","ProductRate"])
newFig2.show()
```

# Bar

```python
newFig=px.bar(data,x="ProductBrand",y="ProductPrice")
newFig.show()
```

# Treemap

```python
fig=px.treemap(data,path=["ProductBrand","ProductName"],values="ProductPrice" ,color='ProductRate', hover_data=['ProductName'],color_continuous_scale='RdBu')
fig.show()
```
